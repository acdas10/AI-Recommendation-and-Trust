```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# -----------------------
# Prepare data
# -----------------------

X = df[[
    "Age",
    "Gender_Male",
    "Gender_Female",
    "Industry_Education",
    "Industry_Student"
]]

X = sm.add_constant(X)

# TRUST MODEL
y_trust = df['AI_Helpfulness_Yes']
model_trust = sm.Logit(y_trust, X).fit(disp=0)

# CONCERN MODEL
y_concern = df['AI_Concern_Yes']
model_concern = sm.Logit(y_concern, X).fit(disp=0)

# -----------------------
# Create prediction data
# Hold controls constant
# -----------------------

mean_age = df["Age"].mean()

predict_df = pd.DataFrame({
    "const": [1,1,1],
    "Age": [mean_age]*3,
    "Gender_Male": [0]*3,
    "Gender_Female": [1]*3,
    "Industry_Education": [0,1,0],
    "Industry_Student": [0,0,1]
})

industry_labels = ["Other Industries", "Education", "Student"]

# -----------------------
# Predicted probabilities
# -----------------------

# TRUST
trust_pred = model_trust.predict(predict_df)

# CONCERN
concern_pred = model_concern.predict(predict_df)

# -----------------------
# Compute Confidence Intervals manually
# -----------------------

def get_ci(model, pred_df):
    params = model.params.values
    cov = model.cov_params().values
    
    # Linear prediction
    lin_pred = np.dot(pred_df.values, params)
    
    # Standard errors
    se = []
    for row in pred_df.values:
        se.append(np.sqrt(np.dot(row.T, np.dot(cov, row))))
    se = np.array(se)
    
    # Convert to probability scale
    lower = 1 / (1 + np.exp(-(lin_pred - 1.96*se)))
    upper = 1 / (1 + np.exp(-(lin_pred + 1.96*se)))
    
    return lower, upper

trust_lower, trust_upper = get_ci(model_trust, predict_df)
concern_lower, concern_upper = get_ci(model_concern, predict_df)

# -----------------------
# Plot
# -----------------------

plt.figure()

# Trust
plt.plot(industry_labels, trust_pred)
plt.fill_between(industry_labels, trust_lower, trust_upper, alpha=0.2)

# Concern
plt.plot(industry_labels, concern_pred)
plt.fill_between(industry_labels, concern_lower, concern_upper, alpha=0.2)

plt.ylim(0,1)
plt.ylabel("Predicted Probability")
plt.title("H1: Adjusted Trust and Concern by Industry")
plt.legend(["Trust", "Concern"])
plt.show()
```

```python
# logistic
import pandas as pd
import statsmodels.api as sm
import numpy as np

# 1. Define variables for H2
# We include Age and Gender as controls to ensure the effect is truly from knowledge
X = df[['AI_Knowledge_Beginner', 'AI_Knowledge_Expert',
       'AI_Knowledge_Intermediate', 'Age', 'Gender_Female']] 
X = sm.add_constant(X)
y = df['AI_Helpfulness_Yes']

# 2. Fit the model
model_h2 = sm.Logit(y, X).fit()

# 3. Calculate Odds Ratios (OR) for easier interpretation
# OR > 1 means the factor increases the likelihood of finding AI helpful
params = model_h2.params
conf = model_h2.conf_int()
conf['OR'] = params
conf.columns = ['Lower CI', 'Upper CI', 'OR']
odds_ratios = np.exp(conf)

print("--- LOGISTIC REGRESSION RESULTS FOR H2 ---")
print(model_h2.summary())
print("\n--- ODDS RATIOS ---")
print(odds_ratios)
```

```python
# Verification with Probit
import statsmodels.api as sm

# 1. Prepare independent variables
# We include 'Intermediate' and 'Expert' to compare them against the 'Beginner' baseline
# Controlling for Age and Gender to ensure Knowledge is the true driver
X_h2_probit = df[['AI_Knowledge_Beginner','AI_Knowledge_Intermediate', 'AI_Knowledge_Expert', 'Age', 'Gender_Female']]
X_h2_probit = sm.add_constant(X_h2_probit)

# 2. Define the dependent variable (H2 focuses on Helpfulness/Trust)
y_h2_probit = df['AI_Helpfulness_Yes']

# 3. Fit the Probit Model
model_h2_probit = sm.Probit(y_h2_probit, X_h2_probit).fit()

print("--- PROBIT VERIFICATION FOR H2 ---")
print(model_h2_probit.summary())
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc
import pandas as pd

# --- STEP 1: DEFINE X AND Y ---
# Using the variables from your H2/H3 analysis
X = df[['AI_Knowledge_Expert', 'AI_Usage_Daily', 'Age', 'Gender_Female']]
X = sm.add_constant(X)
y = df['AI_Helpfulness_Yes']

# --- STEP 2: FIT BOTH MODELS (This defines the names) ---
logit_model = sm.Logit(y, X).fit()
probit_model = sm.Probit(y, X).fit()

# --- STEP 3: GENERATE PREDICTIONS ---
logit_probs = logit_model.predict(X)
probit_probs = probit_model.predict(X)

# --- STEP 4: PLOT ROC CURVE COMPARISON ---
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")

# Logit ROC
fpr_l, tpr_l, _ = roc_curve(y, logit_probs)
plt.plot(fpr_l, tpr_l, label=f'Logit (AUC = {auc(fpr_l, tpr_l):.3f})', color='blue', lw=2)

# Probit ROC
fpr_p, tpr_p, _ = roc_curve(y, probit_probs)
plt.plot(fpr_p, tpr_p, label=f'Probit (AUC = {auc(fpr_p, tpr_p):.3f})', color='red', lw=2, linestyle='--')

plt.plot([0, 1], [0, 1], color='gray', linestyle=':')
plt.title('Performance Comparison: Logit vs. Probit ROC')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# --- STEP 5: PLOT PROBABILITY AGREEMENT ---
plt.figure(figsize=(8, 6))
plt.scatter(logit_probs, probit_probs, alpha=0.4, color='purple')
plt.plot([0, 1], [0, 1], color='black', linestyle='--', label='Identity Line (Perfect Agreement)')
plt.title('Model Consistency: Predicted Probabilities')
plt.xlabel('Logit Probability Estimates')
plt.ylabel('Probit Probability Estimates')
plt.legend()
plt.show()
```
